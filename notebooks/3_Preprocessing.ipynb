# Create Dummies

```python
cat_df = df.select_dtypes(include='object')
```

```python
dummies_df = pd.get_dummies(cat_df)
```

```python
df_merged = pd.merge(dummies_df, df, on=df.index)
df_merged.head()
```

```python
del df_merged['key_0']
```

```python
drop_columns = [col for col in df.select_dtypes(include='object').columns]
drop_columns
```

```python
df_merged.drop(drop_columns, axis=1, inplace=True)
```

# Scale Data

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df_merged)

matrix = scaler.transform(df_merged)

df_final = pd.DataFrame(matrix, columns=df_merged.columns)
```

# Splitting Data

```python
X = df_final.drop('readmitted', axis=1)
y = df_final['readmitted']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```
